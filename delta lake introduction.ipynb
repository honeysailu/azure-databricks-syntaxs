```python
%sql
CREATE TABLE employees_delta (
    emp_id INT,
    emp_name STRING,
    department STRING,
    salary INT,
    updated_date DATE
)
USING DELTA;
```

```python
%sql
DESCRIBE HISTORY default.employees_delta
```

```python
%sql
INSERT INTO employees_delta VALUES
(1, 'Ravi', 'HR', 30000, current_date()),
(2, 'Sita', 'IT', 50000, current_date()),
(3, 'John', 'Finance', 45000, current_date());
```

```python
# View Delta transaction log information using SQL
# For managed tables in serverless, use DESCRIBE HISTORY instead of filesystem access

print("Delta Transaction History:")
print("=" * 60)

# Get transaction history which shows the transaction log activity
history_df = spark.sql("DESCRIBE HISTORY employees_delta")
display(history_df)
```
