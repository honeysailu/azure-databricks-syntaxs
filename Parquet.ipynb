```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

data = [("Sai", "IT", 30000), ("Lakshmi", "HR", 25000)]
columns = ["name", "department", "salary"]

df = spark.createDataFrame(data, columns)

df.show()
```

```python
df.write.mode("overwrite").parquet("/Volumes/workspace/default/myvolume/employees_parquet")
```

```python
df2 = spark.read.parquet("/Volumes/workspace/default/myvolume/employees_parquet")
df2.show()
```

```python
df.write.mode("overwrite").format("parquet").saveAsTable("workspace.default.emp_parquet")
```

```python
df.write.mode("overwrite").format("delta") \
    .saveAsTable("workspace.default.emp_parquet")
```
