```python
from pyspark.sql.functions import trim, col
df_trimmed = df.select([trim(col(c)).alias(c) for c in df.columns])
display(df_trimmed)
df_trimmed.count()
df_trimmed.printSchema()
df_trimmed.columns
display(df_trimmed.describe())
df_trimmed.dropna()
display(df_trimmed.dropDuplicates())
df_trimmed.count()
```

```python
from pyspark.sql.functions import trim, col

# Trim only string columns
df_trimmed = df.select([
    trim(col(c)).alias(c) if dict(df.dtypes)[c] == 'string' else col(c)
    for c in df.columns
])

# Drop nulls
df_clean = df_trimmed.dropna()

# Remove duplicates
df_clean = df_clean.dropDuplicates()

# Show results
display(df_clean)

print("Count:", df_clean.count())
df_clean.printSchema()
display(df_clean.describe())
```

```python
from pyspark.sql.functions import to_date

df_clean = df_clean.withColumn("Date", to_date(col("Date"), "yyyy-MM-dd"))

# df.withColumn(
#     "Date",
#     to_date(col("Date"), "dd-MMM-yy")
# )
display(df_clean)
```

```python
from pyspark.sql.functions import coalesce, to_date, col

df = df.withColumn(
    "Date",
    coalesce(
        to_date(col("Date"), "dd-MMM-yy"),
        to_date(col("Date"), "dd/MM/yyyy"),
        to_date(col("Date"), "yyyy-MM-dd")
    )
) 

display(df)
df.select("Date").show(20, False)
```
